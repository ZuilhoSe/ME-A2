```python
# filtrar colunas
filtered_adh = df_adh[pd.read_csv('dados_adh.txt', header=None)[0].values]
filtered_edu = df_edu[pd.read_csv('dados_edu.txt', header=None)[0].values]
```

```python
# drop all 1991 years adh
filtered_adh = filtered_adh[filtered_adh['ano'] != 1991]

# add all years 2006 from 2022, duplicanting year 2000 if lower than 2010 and 2010 if lower than 2023
novos_dados = []
for i in range(2006, 2023):
    if i not in filtered_adh['ano'].unique():
        if i < 2010:
            dados_replicados = filtered_adh[filtered_adh['ano'] == 2000].copy()
            dados_replicados['ano'] = i
        else:
            dados_replicados = filtered_adh[filtered_adh['ano'] == 2010].copy()
            dados_replicados['ano'] = i
        novos_dados.append(dados_replicados)
filtered_adh = pd.concat([filtered_adh] + novos_dados)

# drop 2000
filtered_adh = filtered_adh[filtered_adh['ano'] != 2000]
```

```python
# porcentagem de valores faltantes e NaN
filtered_adh.isnull().mean() * 100
```

```python
filtered_edu.isnull().mean() * 100
```

```python
merged = pd.merge(filtered_adh, filtered_edu, on=['id_municipio', 'ano'], how='inner')
merged_pib = pd.merge(merged, df_pib, on=['id_municipio', 'ano'], how='inner')
```

```python
merged_pib.isnull().mean() * 100
```

```python
merged_pib.to_csv('data/data.csv', index=False)
```

```python
# Para carregar o dado direto no pandas
df = bd.read_table(dataset_id='br_inep_ideb',
table_id='municipio',
billing_project_id="117003241911")
```

```python
# unique values for each column
df.columns.to_list()
```

```python
print(df['rede'].unique())
print(df['ensino'].unique())
```

```python
df_filtered = df[['ano', 'id_municipio', 'ensino', 'ideb', 'nota_saeb_media_padronizada', 'rede']]

# tirar a media do ideb e da nota_saeb_media_padronizada, por municipio, ano, ensino e rede
df_filtered = df_filtered.groupby(['ano', 'id_municipio', 'ensino', 'rede']).mean().reset_index()

# tirar a media do ideb e da nota_saeb_media_padronizada, por municipio e ano criando uma nova coluna
df_filtered['ideb_media'] = df_filtered.groupby(['id_municipio', 'ano'])['ideb'].transform('mean')

# apagar as colunas ensino, rede, ideb e nota_saeb_media_padronizada
df_filtered = df_filtered.drop(['ensino', 'rede', 'ideb', 'nota_saeb_media_padronizada'], axis=1)

# remover duplicatas
df_filtered = df_filtered.drop_duplicates()
```

```python
# see unique ano
df_filtered['ano'].unique()
```

```python
# add all years from 2006 to 2022 repliating the previous year if not exists
novos_dados = []
for i in range(2006, 2023, 2):
    dados_replicados = df_filtered[df_filtered['ano'] == i - 1].copy()
    dados_replicados['ano'] = i
    novos_dados.append(dados_replicados)

df_filtered = pd.concat([df_filtered] + novos_dados)

print(df_filtered['ano'].unique())
```

```python
merged_pib_ideb = pd.merge(merged_pib, df_filtered, on=['id_municipio', 'ano'], how='inner')

merged_pib_ideb.isnull().mean() * 100
```

```python
# exportar para csv
merged_pib_ideb.to_csv('data/data.csv', index=False)
```
